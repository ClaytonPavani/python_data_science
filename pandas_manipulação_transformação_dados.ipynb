## Questão Exemplo de Como transformar uma coluna com dados STRINGS --> Numéricos

- Float
- INT

```python
# Criação do Data Frame:

# dados da firma
dados1 = {
    'cargo': ['Gerencia', 'Coordenação', 'Supervisão', 'Analista', 'Assistente', 'Operação', 'Auxiliar', 'Estágio', 'Assessoria', 'Consultoria'],
    'quantidade': ['2 pessoas', '1 pessoa', '3 pessoas', '4 pessoas', '5 pessoas', '3 pessoas', '2 pessoas', '1 pessoa', '1 pessoa', '1 pessoa'],
    'salario': ['$10.000 reais', '$8.000 reais', '$7.000 reais', '$5.000 reais', '$4.000 reais', '$3.500 reais', '$3.000 reais', '$1.500 reais', '$2.500 reais', '$6.000 reais']
}

# transformando o dicionário em DataFrame
df = pd.DataFrame(dados1)

# df é o DataFrame com os dados da firma
df
```

```python
# Removendo os caracteres textuais '$ e ,'

# Usando Apply + Função Lambda

df['quantidade'] = df['quantidade'].apply(lambda x: x.replace(' pessoas', '').replace(' pessoa', ''))
df['salario'] = df['salario'].apply(lambda x: x.replace('$', '').replace(' reais', ''))
```

```python
# Após os dados tratados, vamos aplicar as modificações de tipos referentes a cada coluna:

df['quantidade'] = df['quantidade'].astype(np.int64)
df['salario'] = df['salario'].astype(np.float64)
```

```python
# Imprimindo o dataframe tratado:

df
```

## Transformando os dados - Tokenização

- Coluna 'descricao_vizinhaca'

- lower_case
- método replace para caracteres especiais
- método replace para hífens fora de palavras compostas
- método split para finalmente criar TOKENS das palavras:

```python
# Transformamos o texto em letras minúsculas
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.lower()
# Substituimos os caractere especiais
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace('(?<!\w)-(?!\w)', '', regex=True)
# Transformamos o texto em lista, formando o token
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.split()
dados.head()
```

## Atividade prática, converter dados STRINGS para DATETIME:

```python
# Criando o DataFrame com as informações
dados_teste = pd.DataFrame({
    'Data de venda': ['01/01/2022', '05/02/2022', '10/03/2022', '15/04/2022','18/04/2022','20/04/2022'],
    'Valor': [100, 150, 200, 250,80,180]
})

# Exibindo o DataFrame
print(dados_teste)
```

```python
# Exemplo de como converter e formatar uma coluna string --> datetime
# Observe que o formato está Dia / Més / Ano: Por isso usamos o Format

dados_teste['Data de venda'] = pd.to_datetime(dados_teste['Data de venda'], format='%d/%m/%Y')
```
