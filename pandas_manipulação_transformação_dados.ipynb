## QuestÃ£o Exemplo de Como transformar uma coluna com dados STRINGS --> NumÃ©ricos

- Float
- INT

```python
# CriaÃ§Ã£o do Data Frame:

# dados da firma
dados = {
    'cargo': ['Gerencia', 'CoordenaÃ§Ã£o', 'SupervisÃ£o', 'Analista', 'Assistente', 'OperaÃ§Ã£o', 'Auxiliar', 'EstÃ¡gio', 'Assessoria', 'Consultoria'],
    'quantidade': ['2 pessoas', '1 pessoa', '3 pessoas', '4 pessoas', '5 pessoas', '3 pessoas', '2 pessoas', '1 pessoa', '1 pessoa', '1 pessoa'],
    'salario': ['$10.000 reais', '$8.000 reais', '$7.000 reais', '$5.000 reais', '$4.000 reais', '$3.500 reais', '$3.000 reais', '$1.500 reais', '$2.500 reais', '$6.000 reais']
}

# transformando o dicionÃ¡rio em DataFrame
df = pd.DataFrame(dados)

# df Ã© o DataFrame com os dados da firma
df
```

```python
# Removendo os caracteres textuais '$ e ,'

# Usando Apply + FunÃ§Ã£o Lambda

df['quantidade'] = df['quantidade'].apply(lambda x: x.replace(' pessoas', '').replace(' pessoa', ''))
df['salario'] = df['salario'].apply(lambda x: x.replace('$', '').replace(' reais', ''))
```

```python
# ApÃ³s os dados tratados, vamos aplicar as modificaÃ§Ãµes de tipos referentes a cada coluna:

df['quantidade'] = df['quantidade'].astype(np.int64)
df['salario'] = df['salario'].astype(np.float64)
```

```python
# Imprimindo o dataframe tratado:

df
```

## ManipulaÃ§Ã£o de Dados Textuais

- 1Â° - Padronizando coluna 'descriÃ§Ã£o_local'
- Usando LowerCase (minusculo)

```python
# Vamos visualizar nossa coluna de descriÃ§Ã£o local primeiro:

dados['descricao_local']
```

```python
dados['descricao_local'] = dados['descricao_local'].str.lower()
```

```python
dados.head()
```

```python
dados['descricao_local'][3169]
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']',' ',regex=True)
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?!\w)',' ',regex=True)
```

```python
dados['descricao_local'] = dados['descricao_local'].str.split()
dados.head()
```

```python
dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"','',regex=True)
```

```python
dados['comodidades'] = dados['comodidades'].str.split(',')
dados.head()
```
